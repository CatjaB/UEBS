````python
#Task 1: problem 1
#first 10 results of searching
#print out info about each movie (ratings, actors, plots, etc.)
#Group movies into 3 groups
#what is the average length of a movie in each group
#Are longer movies higher rated?

import requests
import pprint as pp

your_api_key = "dd14dc5f" #use mine, or get your own from the above website
page = 1 # each page has 10 results, so page 1 has results 1-10, page 2 has results 11-20, etc
query = "star wars"
result_type="movie" # read the omdb website to see what other options there are
year = "" #leave empty to get all results all years

url = f"http://www.omdbapi.com/?s={query}&page={page}&apikey={your_api_key}&type={result_type}&y={year}"
response = requests.request("GET", url)

star_wars_movies = response.json()['Search'];

star_wars_movies_ids = [ movie['imdbID']
    for movie in star_wars_movies]

#groups for rating
mediocre = []
good = []
amazing = []

#lengths of group rating 
good_length = []
mediocre_length = []
amazing_length = []

print("Here are the first 10 movies: ")
for movie_id in star_wars_movies_ids:
    url = "http://www.omdbapi.com/?i="+movie_id+"&apikey={}".format(your_api_key)
    response_dictionary = requests.request("GET", url).json()
    #get all infos
    titles_string = response_dictionary["Title"]
    titles_list = titles_string.split(", ")
    actors_string = response_dictionary["Actors"]
    actors_list = actors_string.split(", ")
    length_string = response_dictionary["Runtime"]
    length_list = length_string.split(", ")
    plot_string = response_dictionary["Plot"]
    plot_list = plot_string.split(", ")
    year_string = response_dictionary["Year"]
    year_list = year_string.split(", ")
    
    rating_as_number = float(response_dictionary["imdbRating"]) 
   
    print(titles_list)
    print(actors_list)
    print(length_list)
    print(plot_list)
    print(year_list)
    
    #print rating accordingly
    if rating_as_number > 8:
        print("rated amazing")
    elif rating_as_number < 7:
        print("rated mediocre")
    else:
        print ("rated good")
    
    #add movie to one of three groups depending on rating(amazing, medioce, good)
    if rating_as_number > 8:
        amazing.append(movie_id)
        amazing_length.append(int(length_list[0].replace('min', '')))
    elif rating_as_number < 7:
        mediocre.append(movie_id)
        mediocre_length.append(int(length_list[0].replace('min', '')))
    else:
        good.append(movie_id)
        good_length.append(int(length_list[0].replace('min', '')))

#caculate averga
def average(list):
    return sum(list)/len(list)

#give out rated movies, with length and average length.
print("'good' rated movies: ", good, "length: ", good_length)
print("Average Length:", round(average(good_length)))
print("'mediocre' rated movies: ", mediocre, "length: ", mediocre_length)
print("Average Length:", round(average(mediocre_length)))
print("'amazin' rated movies: ", amazing, "length: ", amazing_length)
print("Amazing Length:", round(average(amazing_length)))

    
#The better the movie rating, the shorter the run time. People like short movies. 
````

````python
# Code Examples: get first 10 movies with "star wars" in the title

import requests
import pprint as pp

your_api_key = "dd14dc5f" #use mine, or get your own from the above website
page = 1 # each page has 10 results, so page 1 has results 1-10, page 2 has results 11-20, etc
query = "star wars"
result_type="movie" # read the omdb website to see what other options there are
year = "" #leave empty to get all results all years

url = f"http://www.omdbapi.com/?s={query}&page={page}&apikey={your_api_key}&type={result_type}&y={year}"
response = requests.request("GET", url)

pp.pprint(response.json())

````

### Search forecast for a location:

`

````python
import requests
import pprint as pp
#example query

api_key = "2aee514fa2e3493e8b5100342202610" 
# this is a temporary api key - please go to the above website and create your own.

city = "Edinburgh"
days = 3
api_url_current_weather = f"http://api.weatherapi.com/v1/forecast.json?key={api_key}&q={city}&days={days}"

response = requests.request("GET", api_url_current_weather)
weather_edinburgh = response.json()
````
